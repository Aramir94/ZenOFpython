# Generators

- 함수를 실행할때 실행되는것이 아니라 객체를 보내준다
- 제너레이터는 객체를 next() 로 불러야 실행된다 
- 다실행이 되었을때는 stopiteration이 불러와진다

```python
def countdown(num):
    print('Starting')
    while num > 0:
        yield num
        num -= 1

# this will not print 'Starting'
cd = countdown(3)

# this will print 'Starting' and the first value
print(next(cd))

# will print the next values
print(next(cd))
print(next(cd))

# this will raise a StopIteration
print(next(cd))
```

```python
# you can iterate over a generator object with a for in loop
cd = countdown(3)
for x in cd:
    print(x)
```

```python
# you can use it for functions that take iterables as input
cd = countdown(3)
sum_cd = sum(cd)
print(sum_cd)

cd = countdown(3)
sorted_cd = sorted(cd)
print(sorted_cd)
```

# Generators save memroy!
- generator는 lazily, 필요할때만 사용하면 되기에 메모리를 많이 줄일수 있다
- 큰 데이터 프레임 쓸떄는 더 
- 모든 요소를 생성하고 쓰는게 아니기 떄문에 더 메모리를 줄여준다

```python
# without a generator, the complete sequence has to be stored here in a list
def firstn(n):
    num, nums = 0, []
    while num < n:
        nums.append(num)
        num += 1
    return nums

sum_of_first_n = sum(firstn(1000000))
print(sum_of_first_n)

import sys
print(sys.getsizeof(firstn(1000000)), "bytes")
```

```python
# with a generator, no additional sequence is needed to store the numbers
def firstn(n):
    num = 0
    while num < n:
        yield num
        num += 1

sum_of_first_n = sum(firstn(1000000))
print(sum_of_first_n)
import sys
print(sys.getsizeof(firstn(1000000)), "bytes")
```

# 피보나치 예제

```python
def fibonacci(limit):
    a, b = 0, 1 # first two fibonacci numbers
    while a < limit:
        yield a
        a, b = b, a + b

fib = fibonacci(30)
# generator objects can be converted to a list (only used for printing here)
print(list(fib))
```

# Generator expressions

- generator가 list comprehensions 보다는 느리다 
- https://stackoverflow.com/questions/11964130/list-comprehension-vs-generator-expressions-weird-timeit-results/11964478#11964478 

```python
# generator expression
mygenerator = (i for i in range(1000) if i % 2 == 0)
print(sys.getsizeof(mygenerator))

# list comprehension
mylist = [i for i in range(1000) if i % 2 == 0]
print(sys.getsizeof(mylist))
```

# Generator 컨셉

- generator은 크게 보면 iterable objectdlek
- yeild 코드를 class로 제작하게 되면

```python
class firstn:
    def __init__(self, n):
        self.n = n
        self.num = 0
        
    def __iter__(self): # 이터하게 돌려준다
        return self
    
    def __next__(self): # 넥스트
        if self.num < self.n:
            cur = self.num
            self.num += 1
            return cur
        else:
            raise StopIteration()
             
firstn_object = firstn(1000000)
print(sum(firstn_object))
```
